## 3. ***Data Wrangling***

### Data Wrangling Code

# Titles Dataset

````python
# Remove duplicate rows
# Remove Duplicate Values - Titles Dataset
titles_df = titles_df.drop_duplicates()
print("Number of Duplicates in the titles dataset", titles_df.duplicated().sum())
````

````python
titles_df.isnull().sum()
````

# Missing Values Analysis(Titles Dataset)
 From the analysis of missing values in the titles dataset, the following observations are made:
  
*   **decription** : has 199 missing values, which is relatively small and can be ignored or filled with a placeholder.
*  **age_certification** : has 6468 missing values, which is relatively small and can be ignored or filled with a placeholder.
*  **seasons** : has 8511 missing values, which is expected since this column is only applicable to TV shows and not movies.
* **imdb_id** : has 667 missing values, suggesting incomplete linkage with IMDB records.
* **imdb_score**  and **imdb_votes** :have over 1000 missing values, which may impact rating-based analysis.
*  **tmdb_popularity** and **tmdb_score** also contain missing values, though comparatively fewer.  







````python
# Write your code to make your dataset analysis ready.

# Handle missing values in titles dataset

# Fill missing descriptions
titles_df.loc[:, "description"] = titles_df.loc[:, "description"].fillna(
    "No overview found."
)

# Fill missing age certifications
titles_df.loc[:, "age_certification"] = titles_df.loc[:, "age_certification"].fillna(
    "Unknown"
)

# Fill missing seasons with 0 because movies do not have seasons
titles_df.loc[:, "seasons"] = titles_df.loc[:, "seasons"].fillna(
    0
)
titles_df.loc[:, 'seasons'] = titles_df.loc[:, 'seasons'].astype(int)

# Fill missing IMDb score using mean value
titles_df.loc[:, "imdb_score"] = titles_df.loc[:, "imdb_score"].fillna(
    titles_df.loc[:, "imdb_score"].mean()
)

# Fill missing IMDb votes using median value
titles_df.loc[:, "imdb_votes"] = titles_df.loc[:, "imdb_votes"].fillna(
    titles_df.loc[:, "imdb_votes"].median()
)

# Fill missing TMDB popularity using median value
titles_df.loc[:, "tmdb_popularity"] = titles_df.loc[:, "tmdb_popularity"].fillna(
    titles_df.loc[:, "tmdb_popularity"].median()
)

# Fill missing TMDB score using median value
titles_df.loc[:, "tmdb_score"] = titles_df.loc[:, "tmdb_score"].fillna(
    titles_df.loc[:, "tmdb_score"].mean()
)

# Check whether missing values are removed
titles_df.isnull().sum()
````

# Credits Dataset

````python
# Remove Duplicate Values - Credits Dataset
credits_df = credits_df.drop_duplicates()
credits_df.duplicated().sum()

# Handle missing values in credits dataset

# Fill missing character names with "Unknown" because directors and some cast records do not have character names
credits_df.loc[:, "character"] = credits_df.loc[:, "character"].fillna(
    "Unknown"
)

# Check whether missing values are removed
credits_df.isnull().sum()
````

````python
# Chart - 8 visualization code
#Average IMDB score by Top Directors (Bivariate)
# Merge titles and credits datasets using the common 'id' column
df = pd.merge(titles_df, credits_df, on='id', how="left")

# Filter only director records from the merged dataset
directors = df[df["role"] == "DIRECTOR"]

# Select directors who have worked on at least 5 titles
top_directors = directors.groupby('name').filter(
    lambda x: len(x) >= 5
)

# Calculate average IMDb score for each director
director_ratings = top_directors.groupby('name')['imdb_score'].mean()

# Sort directors by average IMDb score and select top 10
director_ratings = director_ratings.sort_values(
    ascending=False
).head(10)

# Create figure
plt.figure(figsize=(8,5))

# Create bar chart for top directors by average IMDb score
sns.barplot(
    x=director_ratings.values,
    y=director_ratings.index
)

# Add chart title
plt.title("Top Directors by Average IMDb Score")

# Add x-axis label
plt.xlabel("Average IMDb Score")

# Add y-axis label
plt.ylabel("Director Name")

# Display chart
plt.show()
````
